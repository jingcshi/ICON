```python
Nitems_recorded = 20
data = {'ID':[], 'name': []}
for i in range(1,Nitems_recorded+1):
    data[f'item{i}'] = []
for n in taxo.nodes():
    data['ID'].append(n)
    data['name'].append(taxo.get_label(n))
    cat_items = itemdf[itemdf['catid'] == n].reset_index(drop=True)
    nitems = min(len(cat_items), Nitems_recorded)
    for j in range(1, nitems+1):
        data[f'item{j}'].append(cat_items.loc[j-1]['itemtitle'])
    for j in range(nitems+1, Nitems_recorded+1):
        data[f'item{j}'].append(None)
df = pd.DataFrame(data)
```

```python
df.to_csv("./ebay_us_20item.csv", index=False)
```
